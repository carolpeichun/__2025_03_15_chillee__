- 這裡是建立`正弦波`的圖表範例

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸數值，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2*np.pi, 0.1)

# 計算 y 軸數值，使用 sin() 函數
y = np.sin(x)

# 繪製圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('sin plot')
plt.xlabel('x')
plt.ylabel('sin(x)')

# 顯示圖表
plt.show()

import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸數值，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2*np.pi, 0.1)

# 計算 y 軸數值，使用 cos() 函數
y = np.cos(x)

# 繪製圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('餘弦波圖表')
plt.xlabel('x')
plt.ylabel('cos(x)')

# 顯示圖表
plt.show()
```
